```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
salarios_marcos = np.array([100,200,400,300])
salarios_gileno = np.array([50,300,500,450])
```

```python
plt.plot(salarios_marcos, c='Black', ls='--', marker='s', ms=8, label="Selários Marcos")
plt.plot(salarios_gileno, c='Red', ls='--', marker='s', ms=8, label="Selários Gileno")
plt.show()
```

## colocando legenda

```python
plt.plot(salarios_marcos, c='Black', ls='--', marker='s', ms=8, label="Selários Marcos")
plt.plot(salarios_gileno, c='Red', ls='--', marker='^', ms=8, label="Selários Gileno")
plt.legend()
plt.show()
```

```python
plt.plot(salarios_marcos, c='Black', ls='--', marker='s', ms=8, label="Selários Marcos")
plt.plot(salarios_gileno, c='Red', ls='--', marker='^', ms=8, label="Selários Gileno")
plt.legend(loc='upper left')
plt.show()
```

```python
plt.plot(salarios_marcos, c='Black', ls='--', marker='s', ms=8, label="Selários Marcos")
plt.plot(salarios_gileno, c='Red', ls='--', marker='^', ms=8, label="Selários Gileno")
plt.legend(loc='upper right')
plt.show()
```

```python
salarios_pedro = [200, 150, 300, 700]
plt.plot(salarios_marcos, c='Black', ls='--', marker='s', ms=8, label="Selários Marcos")
plt.plot(salarios_gileno, c='Red', ls='--', marker='^', ms=8, label="Selários Gileno")
plt.plot(salarios_pedro, c='Green', ls='--', marker='o', ms=8, label="Selários Pedro")
plt.legend()
plt.show()
```
