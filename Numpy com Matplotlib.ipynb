```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
salario = np.array([100, 200, 300, 500, 400, 150])
salario
```

```python
plt.plot(salario)
```

```python
plt.plot(salario, c='Black')
```

```python
plt.plot(salario, c='Black', ls='--')
```

```python
plt.plot(salario, ls='--', marker='s')
```
